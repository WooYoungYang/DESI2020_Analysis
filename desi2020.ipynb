```python
import pandas as pd
import numpy as np
```

```python
desi20_o = pd.read_csv('DESI_2020.csv', encoding='utf-8')
desi20 = desi20_o.copy()
desi20
```

```python
desi20['indicator'].unique()
```

```python
desi20 = desi20.drop(columns=['observation',
                              'unit_measure',
                              'flag',
                              'note'])
desi20.head(3)
```

```python
desi20a = desi20[(desi20['indicator']=='desi')]
desi20a = desi20a.drop(columns=['indicator'])
desi20a = desi20a.rename(columns={'time_period':'Year',
                                  'ref_area':'Country Code',
                                  'breakdown':'Indicator',
                                  'value':'Weighted Score'})
desi20a
```

```python
desi20a['Indicator'].unique()
```

```python
desi20a.loc[desi20a['Indicator']=='desi_1_conn', 'Indicator'] = '1. Connectivity'
desi20a.loc[desi20a['Indicator']=='desi_2_hc', 'Indicator'] = '2. Human Capital'
desi20a.loc[desi20a['Indicator']=='desi_3_ui', 'Indicator'] = '3. Use of Internet'
desi20a.loc[desi20a['Indicator']=='desi_4_idt', 'Indicator'] = '4. Integration of Digital Technology'
desi20a.loc[desi20a['Indicator']=='desi_5_dps', 'Indicator'] = '5. Digital Public Services'
desi20a
```

```python
desi_sum = desi20a.groupby(['Country Code', 'Year']).sum()
desi_sum = desi_sum.reset_index()
desi_sum['Indicator'] = 'Overall'
desi_sum
```

```python
desi20b = pd.concat([desi20a, desi_sum])
desi20b
```

```python
desi_rank = desi20b.groupby(['Year', 'Indicator'])['Weighted Score'].rank(ascending=False)
desi20b['Rank'] = desi_rank.astype('int8')
desi20b
```

```python
cc = pd.read_csv('ISO3166-1_alpha-2.csv', encoding='utf-8')
cc.head(3)
```

```python
cc = cc.rename(columns={'Entry':'Country Code',
                        'Country name (using title case)':'Country'})
cc.head(3)
```

```python
desi20c = desi20b.merge(cc, how='left', on='Country Code')
desi20c
```

```python
desi20c['Country'].unique()
```

```python
desi20c[desi20c.isnull().any(axis=1)]
```

```python
desi20c[desi20c.isnull().any(axis=1)]['Country Code'].unique()
```

```python
desi20c.loc[desi20c['Country Code']=='EL', 'Country'] = 'Greece'
desi20c.loc[desi20c['Country Code']=='EU', 'Country'] = 'EU 28'
desi20c.loc[desi20c['Country Code']=='UK', 'Country'] = 'United Kingdom'
desi20c.loc[desi20c['Country Code']=='CZ', 'Country'] = 'Czech Republic'
desi20c
```

```python
collist = desi20c.columns.tolist()
collist = collist[0:2] + collist[-1:] + collist[2:-1]
collist
```

```python
desi20c = desi20c[collist]
desi20c
```

```python
pp = pd.read_csv('WPP2019_TotalPopulationBySex.csv', encoding='utf-8')
pp.head(3)
```

```python
pp = pp.drop(columns=['LocID',
                      'VarID',
                      'Variant',
                      'MidPeriod',
                      'PopMale',
                      'PopFemale',
                      'PopDensity'])
pp = pp.rename(columns={'Location':'Country',
                        'Time':'Year',
                        'PopTotal':'Population'})
#pp = pp.rename(columns={'PopTotal':'Population'})
pp['Population'] = pp['Population'] * 1000
pp['Population'] = pp['Population'].astype('int64')
pp.loc[pp['Country']=='Czechia', 'Country'] = 'Czech Republic'
#pp.loc[pp['Location']=='Czechia', 'Location'] = 'Czech Republic'
pp.head(3)
```

```python
pp.info()
```

```python
desi20d = desi20c.merge(pp,
                        how='left',
                        left_on=['Year', 'Country'],
                        right_on=['Year', 'Country'])
                        #right_on=['Time', 'Location'],
                        #validate='many_to_one')
desi20d
```
